Data Preprocessing in Machine Learning Using Pandas
Data Preprocessing is required tasks fro cleaning the data and making it suitable for a machine learning model which also increase the accuracy and efficency of a machine learning model.

 Pandas is an oen source Python package that is most widely used for data science/data analaysis and machine learning tasks. It is built on top of another package named Numpy, which provides support for multi-dimensional arrays. Preprocessing is the process of doing a pre-analysis of data, in order to transform them into a standard and normalized format.

 Preprocessing involves the following aspects:
 - missing values
 - data standardization
 - data normalization
 - data minning

```python
# replace single quote with blank and make the date formatted.
df["Date"] =df["Date"].replace("'", "", regex = True)
df
```

```python
# Handle null values, return dataframe with no empty cells
# We can use df.dropna(), it removes empty cells but will not change the dataframe
# If we want data frame to be changed, simply use inplace=True.
new_df=df.dropna()
new_df
```

```python
isnull_count_newdf = new_df.isnull().sum()
print(isnull_count_newdf)
```

```python
#another way of handling the null values is by replacing the null values.
df.fillna(130, inplace=True)
df
```
